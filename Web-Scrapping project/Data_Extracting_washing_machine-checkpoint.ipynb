## OBJECTIVE

The Objective of this project is to analyze the pricing patterns of washing machines available on Flipkart using webscaping, Exploratory Data Analysis (EDA), and data visualization techniques. The goal is to identify key factors influencing product prices and get insights that would help the customer make well informed purchase decisions.

Target Feature: 
- Price

Independent Features :
- Brand
- Capacity
- Load_type
- Spin_speed 
- Machine_type
- Motor_warranty
- Discount
- Ratings
- Total_ratings
- Total_reviews

```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import numpy as np
```

```python
Brand = []
Capacity = []
Load_type =[]
Spin_speed = []
Machine_type = []
Motor_warranty = []
Offer_price = []
Original_price = []
Discount = []
Ratings = []
Quality = []
```

```python
for i in range(1,50):
    url = f'https://www.flipkart.com/search?q=washing+machine&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}'
    response = requests.get(url)
    soup = bs(response.text)
    section = soup.find_all('div',class_ = 'tUxRFH')
    for i in section:
        t = i.find('div',class_ = 'KzDlHZ')
        text  = t.text

        #Brands
        brands = re.findall(r'^\w+ ',text)
        if brands:
            Brand.append(brands[0])
        else:
            Brand.append("Unknown")

        #Capacity
        cap = re.findall(r'\d+ kg',text)
        if cap:
            Capacity.append(cap[0])
        else:
            Capacity.append(np.nan)

        #Load
        load = re.findall(r'\w+ Load',text)
        if load:
            Load_type.append(load[0])
        else:
            Load_type.append(np.nan)

        #machine_type
        mt = re.findall(r'(Automatic|Semi Automatic)',text)
        if mt:
            Machine_type.append(mt[0])
        else:
            Machine_type.append(np.nan)

    for i in section:
        t = i.find('div',class_ = '_6NESgJ')
        text1 = t.text

        #Spin
        spin = re.findall(r'\d+ (?:RPM|rpm)',text1)
        if spin:
            Spin_speed.append(spin[0])
        else:
            Spin_speed.append(np.nan)

        #Motor_warranty
        mw = re.findall(r'\d+ \w+ (?:Warranty|warranty)',text1)
        if mw:
            Motor_warranty.append(mw[0])
        else:
            Motor_warranty.append(np.nan)

        #offer_price
        op = i.find('div',class_ = 'Nx9bqj _4b5DiR')
        if op:
            Offer_price.append(op.text)
        else:
            Offer_price.append(np.nan)

        #OG_price
        ogp = i.find('div',class_ = 'yRaY8j ZYYwLA')
        if ogp:
            Original_price.append(ogp.text)
        else:
            Original_price.append(np.nan)

        #discount
        disc = i.find('div',class_ = 'UkUFwK')
        if disc:
            Discount.append(disc.text)
        else:
            Discount.append(np.nan)

        #ratings
        rat = i.find('div',class_ = 'XQDdHH')
        if rat:
            Ratings.append(rat.text)
        else:
            Ratings.append(np.nan)

        #no_of_ratings & reviews --->quality
        quality = i.find('span',class_='Wphh3N')
        if quality:
            Quality.append(quality.text)
        else:
            Quality.append(np.nan)
```

```python
Data = {
    'Brand':Brand,
    'Capacity' : Capacity,
    'Load_type' : Load_type,
    'Spin_speed' : Spin_speed,
    'Machine_type' : Machine_type,
    'Motor_warranty' : Motor_warranty,
    'Offer_price' : Offer_price,
    'Original_price' : Original_price,
    'Discount' : Discount,
    'Ratings' : Ratings,
    'Quality' : Quality
}
```

```python
df = pd.DataFrame(Data)
df
```

```python
df.to_csv('Washing_machine2.csv',index = False)
```
