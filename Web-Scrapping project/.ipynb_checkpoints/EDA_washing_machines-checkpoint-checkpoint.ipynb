```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('washing_machine_final1.csv')
df
```

## Univariate Analysis 

```python
numerical_columns = df.select_dtypes(include = ['int64' , 'float64'])
categorical_columns = df.select_dtypes(include = ['object'])
```

```python
# discrete data statistical non-visual analysis
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(categorical_columns)
```

```python
# Numerical data statistical non-visual analysis
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(numerical_columns)
```

```python
# IQR
def iqr(df,column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = IQR - 1.5 * Q1
    upper = IQR + 1.5 * Q3
    outlier = df.loc[(df[column] < lower) | (df[column] > upper),column]
    print(len(outlier))

    plt.figure(figsize = (9,7))
    df[column].plot(kind='box')
    plt.show()

    return outlier
```

```python
iqr(df,'Offer_price')
```

# Project Conclusion:

### Discount:

Discount shows a weak negative correlation with Offer Price (-0.26), indicating that higher discounts are slightly associated with lower prices.

### Capacity and Price:

Capacity has a moderate positive correlation with Offer Price (0.25) and Original Price (0.27). This suggests that larger washing machines tend to be more expensive.

### Spin Speed:

Spin Speed has weak correlations with most variables but has a slightly higher correlation with Discount (0.16) and Capacity (0.19), indicating faster spin speed machines may have slightly better discounts and larger capacities.

### Motor Warranty:

Motor Warranty shows weak to moderate correlations with several features:
Positive correlation with Offer Price (0.20), suggesting machines with longer motor warranties tend to be slightly more expensive.
A small positive correlation with Capacity (0.026), suggesting machines with larger capacities tend to have better motor warranties.

## Load Type and Machine Type Impact on price
Load Type and Machine Type have a clear impact on the price. Consumers opting for Front Load or Automatic machines are likely to pay a premium for better performance, energy efficiency, and convenience.

Top Load and Semi-Automatic machines are more affordable, making them more accessible to price-sensitive customers.
