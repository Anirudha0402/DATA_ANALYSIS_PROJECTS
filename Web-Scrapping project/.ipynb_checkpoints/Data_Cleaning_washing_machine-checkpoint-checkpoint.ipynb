```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('Washing_machine2.csv')
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

# DATA CLEANING

```python
df.isna().sum()
```

```python
# correct values in column
df['Brand'] = df['Brand'].str.upper()
df['Capacity'] = df['Capacity'].replace('kg','',regex=True)
df['Capacity'] = pd.to_numeric(df['Capacity'], errors='coerce')
df['Capacity'] = df['Capacity'].fillna(df['Capacity'].mean()).astype(int)
df['Load_type'] = df['Load_type'].fillna(df['Load_type'].mode()[0]).astype(object)
df['Spin_speed'] = df['Spin_speed'].replace(r'\D','',regex=True).astype(int)
df['Machine_type'] = df['Machine_type'].fillna(df['Machine_type'].mode()[0])
df['Motor_warranty'] = df['Motor_warranty'].replace(r'\D','',regex=True).astype(float)
df['Offer_price'] = df['Offer_price'].replace(r'\D','',regex=True)
df['Original_price'] = df['Original_price'].replace(r'\D','',regex=True)
df['Discount'] = df['Discount'].replace(r'\D','',regex=True)
df['Ratings'] = df['Ratings'].fillna(df['Ratings'].mean()).round(1)
df['Total_ratings'] = df['Quality'].str.split('&').str[0]
df['Total_reviews'] = df['Quality'].str.split('&').str[1]
#correcting the values in column
df['Motor_warranty'] = df['Motor_warranty'].fillna(df['Motor_warranty'].mean()).astype(int)
df['Offer_price'] = pd.to_numeric(df['Offer_price'],errors= 'coerce')
df['Offer_price'] = df['Offer_price'].fillna(df['Offer_price'].mean()).astype(int)
df['Original_price'] = pd.to_numeric(df['Original_price'], errors = 'coerce')
df['Original_price'] = df['Original_price'].fillna(df['Original_price'].mean()).astype(int)
df['Discount'] = pd.to_numeric(df['Discount'],errors = 'coerce')
df['Discount'] = df['Discount'].fillna(df['Discount'].mean()).astype(int)
df['Total_ratings'] = df['Total_ratings'].replace(r'\D','',regex=True)
df['Total_ratings'] = pd.to_numeric(df['Total_ratings'],errors = 'coerce')
df['Total_ratings'] = df['Total_ratings'].fillna(df['Total_ratings'].mean()).astype(int)
df['Total_reviews'] = df['Total_reviews'].replace(r'\D','',regex = True)
df['Total_reviews'] = pd.to_numeric(df['Total_reviews'],errors = 'coerce')
df['Total_reviews'] = df['Total_reviews'].fillna(df['Total_reviews'].mean()).astype(int)
df = df.drop(columns = ['Quality'],axis = 1)
df
```

```python
df.isna().sum()
```

```python
df.to_csv('washing_machine_final1.csv',index = False)
```
