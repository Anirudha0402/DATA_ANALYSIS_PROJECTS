```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

# Data Cleaning

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing reviews_per_month with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0, inplace=True)
```

```python
# Drop rows with missing host_name or name
df.dropna(subset=['name', 'host_name'], inplace=True)
```

```python
# Remove outliers: price > 1000 is extreme
df = df[df['price'] <= 1000]
```

```python
# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df.shape
```

# Univariate Analysis

```python
# Price distribution
plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=50, kde=False, color='blue')
plt.title("Price Distribution")
plt.xlim(0,1000)
plt.show()
```

```python
# Room type distribution
plt.figure(figsize=(6,4))
sns.countplot(x='room_type', data=df)
plt.title("Room Type Distribution")
plt.show()
```

# Bivariate Analysis

```python
# Price by room type
plt.figure(figsize=(8,5))
sns.boxplot(x='room_type', y='price', data=df)
plt.ylim(0,500)
plt.title("Price by Room Type")
plt.show()
```

```python
# Average price by neighborhood
neighborhood_price = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
neighborhood_price.plot(kind='bar', color='orange')
plt.title("Average Price by Neighbourhood Group")
plt.ylabel("Avg Price")
plt.show()
```

```python
# Correlation heatmap (numeric features)
plt.figure(figsize=(8,6))
sns.heatmap(df[['price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
